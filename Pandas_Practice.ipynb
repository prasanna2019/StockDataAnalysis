```python
import pandas as pd
import matplotlib as mtb
```

```python
data= {
    "names":['Prasad', 'Ajay', 'Ravi', 'Sujay', 'Shreya'],
    "id":[1,2,3,4,5],
    "ages":[21, 25, 16, 19, 35]
}
salary={
    "id":[1,2,3,4,5, 6],
    "amount" : [200, 300, 310, 120, 233, 199]
}

```

```python
df= pd.DataFrame(data)
df
```

```python
df_salary= pd.DataFrame(salary)
df_salary
```

```python
df['Adult']= df['ages'].apply(adultCheck)
```
